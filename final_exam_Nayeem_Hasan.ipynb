```python
# Section 4 - Manual Implementation
# Only Week 4–6 techniques used (lists, loops, basic functions, basic input validation)

tasks = []


def display_menu():
    """
    Display menu options (Week 3 formatting practice).

    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Quit
    """
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Quit")


def get_user_choice():
    """
    Get and validate user choice (Week 6 input validation).

    >>> # Requires manual input
    >>> get_user_choice()  # doctest: +SKIP
    """
    choice = input("Enter choice (1-4): ").strip()
    if choice in ["1", "2", "3", "4"]:
        return choice
    else:
        print("Invalid choice. Please enter 1–4.")
        return None


def add_task():
    """
    Add a task to the global tasks list.

    >>> tasks.clear()
    >>> add_task()  # doctest: +SKIP
    """
    text = input("Enter task: ").strip()
    if text == "":
        print("Task cannot be empty.")
    else:
        tasks.append(text)
        print(f"Added: {text}")


def list_tasks():
    """
    List all tasks (Week 4 loops).

    >>> tasks.clear()
    >>> tasks.extend(["A", "B"])
    >>> list_tasks()  # doctest: +SKIP
    1: A
    2: B
    """
    if not tasks:
        print("No tasks available.")
    else:
        for i, t in enumerate(tasks):
            print(f"{i+1}: {t}")


def remove_task():
    """
    Remove a task by index (Week 6 defensive checks).

    >>> tasks.clear()
    >>> tasks.extend(["A", "B", "C"])
    >>> remove_task()  # doctest: +SKIP
    """
    if not tasks:
        print("No tasks to remove.")
        return

    index_text = input("Enter task number to remove: ").strip()
    try:
        index = int(index_text)
        index0 = index - 1
        if 0 <= index0 < len(tasks):
            removed = tasks[index0]
            del tasks[index0]
            print(f"Removed: {removed}")
        else:
            print("Invalid index.")
    except:
        print("Please enter a number.")


def main():
    """
    Main programme loop (Week 5 loop patterns).
    >>> main()  # doctest: +SKIP
    """
    while True:
        display_menu()
        choice = get_user_choice()
        if choice == "1":
            add_task()
        elif choice == "2":
            list_tasks()
        elif choice == "3":
            remove_task()
        elif choice == "4":
            print("Goodbye!")
            break


if __name__ == "__main__":
    import doctest
    doctest.testmod()
main()
```
